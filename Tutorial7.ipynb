```python
import math

import numpy as np

import networkx as nx

import matplotlib.pyplot as plt
```

```python
# Define the graph with nodes and edges
edges = [
    (1, 2), (1, 3), (1,4), (2, 3), (2, 4),
    (3, 4)
]

# Initialize the graph
G = nx.Graph()
for edge in edges:
    G.add_edge(*edge)

# Define the layout for visualization
pos = nx.spring_layout(G, seed=42)  # spring layout for a clear view

# Plot the graph
plt.figure(figsize=(12, 12))  # Set the figure size for larger visualization
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray",
        node_size=1000, font_size=16)
plt.title("Graph with Edges", fontsize=18)
plt.show()
```

**Part B**

```python
# Define the graph with nodes and edges
edges = [
    (0, 1, 7), (0, 3, 9), (0, 5, 14), (1, 2, 10), (1, 3, 15),
    (2, 3, 11), (2, 4, 2), (3, 4, 6), (3, 5, 9), (4, 6, 7),
    (5, 6, 10), (5, 7, 12), (6, 8, 8), (7, 8, 5), (7, 9, 6), (8, 9, 3)
]

# Initialize the graph
G = nx.Graph()
G.add_weighted_edges_from(edges)

# Define the layout for visualization
pos = nx.spring_layout(G, seed=42)  # spring layout for a clear view

# Plot the graph
plt.figure(figsize=(12, 12))  # Set the figure size for larger visualization
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray",
        node_size=1000, font_size=16)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): w for u, v, w in edges}, font_size=12)
plt.title("Graph with Weighted Edges", fontsize=18)
plt.show()
```

```python
print(nx.dijkstra_path(G, 1, 9))
```

```python
T = nx.minimum_spanning_tree(G)
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color="lightblue", node_size=500)
nx.draw_networkx_edges(G, pos, edge_color="grey")
nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")
nx.draw_networkx_edge_labels(
    G, pos, edge_labels={(u, v): d["weight"] for u, v, d in G.edges(data=True)}
)
nx.draw_networkx_edges(T, pos, edge_color="green", width=2)
plt.axis("off")
plt.show()
```
